## How student habits correlate with academic performance, a brief analysis

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
sns.set_theme(style="darkgrid")
```

```python
df = pd.read_csv('student_habits_performance.csv')
```

```python
sns.pairplot(data=df)
```

```python
numerical_cols = [ 
'study_hours_per_day',
'social_media_hours',
'netflix_hours',
'attendance_percentage',
'sleep_hours',
'exam_score']
```

```python
sns.heatmap(data=df[numerical_cols].corr(), annot=True, cmap='magma')
sns.pairplot(data=df[numerical_cols], hue='exam_score', palette='magma')
```

```python
sns.violinplot(data=df, y='exam_score', x='gender', palette='magma')
```

```python
sns.lmplot(data=df, x='study_hours_per_day', y='exam_score', hue='gender', palette='magma', height=6, aspect=1.5)
```

```python
sns.violinplot(data=df, x='exercise_frequency', y='exam_score', palette='magma')
```

```python
sns.histplot(df['exam_score'], bins=20, kde=True, color='purple')
```

```python
g = sns.FacetGrid(df, col='mental_health_rating')
g.map(sns.histplot, 'exam_score',color='purple', kde=True)
```

```python
sns.scatterplot(data=df, x='study_hours_per_day', y='exam_score', hue='mental_health_rating', palette='viridis')
```
